# Lendo os dados

## Insights dos arquivos json
- Estão em formato json, como se cada linha fosse um registro.
- Modos de leitura:
    - Leitura por entrada: simula mais a condição real de aquisição, mas é mais lenta.
    - Leitura por bloco: requer ter todos os dados já inseridos, é mais rápido.
- Estrutura de dados:
    - SQL: devido a grande quantidade de dados é interessante por manter os arquivos no HD e armazenar na RAM sob demanda. É ideal se fosse registro em streamming.
    - Pandas: Mantêm todos os arquivos no RAM, mas facilita leitura em bloco completo.
- Decisão: já que os dados estão disponibilizados será usado o pandas para ler e formar o banco de dados SQL, por conveniência e praticidade. Já que os dados já estarão em memória RAM, será apresentado a solução usando linguagem SQL e o pandas, para demonstrar competência com as duas ferramentas.

```python
%%time
# Reading and storing all data in single DataFrame
data_list = []
for i in range(2009,2013):
    with open(os.path.join('.','data',f'data-sample_data-nyctaxi-trips-{i}-json_corrigido.json')) as f:
        file_content = f.read()
        #Brackets inserted to read JSON as list
        file_content = '[' + file_content.replace('\n',',')[:-1] + ']'
    json_data = json.loads(file_content)
    data_list.append(pd.DataFrame(json_data))
    
data = data_list[0]
for i in range(1,len(data_list)):
    data = data.append(data_list[i], ignore_index = True)
```

```python
data.head()
```

```python
data.info()
```

```python
#Transforming date Series from string to datetime types
data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'])
data['dropoff_datetime'] = pd.to_datetime(data['dropoff_datetime'])
```

```python
data.info()
```

```python
# Checking pandas table schema
from pandas.io.json import build_table_schema
build_table_schema(data)
```

```python
try:  
    con = sqlite3.connect(os.path.join('.','data','NYC_taxi.db'))
    data.to_sql("ride_hist", con)
except sqlite3.Error as error:
    print("Error while connecting to sqlite", error)
finally:
    con.close()
```

## Quesito mínimo 3

Faça um histograma da distribuição mensal, nos 4 anos, de corridas pagas em dinheiro;

### In SQL:

```python
# Querying information
statement = '''
SELECT strftime("%m", pickup_datetime) AS Month, COUNT('vendor_id') AS num_of_cash_transactions
FROM ride_hist
WHERE 
payment_type IN (SELECT payment_type FROM payment_lookup WHERE payment_lookup == 'Cash')
GROUP BY strftime("%m", pickup_datetime)
'''

month_cash_pays, description = fetch_database(os.path.join('.','data','NYC_taxi.db'), statement)

print(month_cash_pays)
```

```python
# Plotting results
plt.figure()
x = [k for k,v in month_cash_pays]
y = [v for k,v in month_cash_pays]
plt.bar(x,y)
plt.xticks(range(0,12), [calendar.month_abbr[i] for i in range(1,13)])
plt.grid()
plt.show()
```

#### Extra

```python
statement = '''
SELECT strftime("%Y", pickup_datetime), strftime("%m", pickup_datetime), COUNT('vendor_id'), lower(payment_type)
FROM ride_hist
GROUP BY strftime("%Y-%m", pickup_datetime) || lower(payment_type)
'''

month_cash_pays, description = fetch_database(os.path.join('.','data','NYC_taxi.db'), statement)

df = pd.DataFrame(data = month_cash_pays, columns=['year','month','count','payment_type'])

g = sns.FacetGrid(df, row='year', hue='payment_type',height=4, aspect=2)
g.map(sns.barplot, 'count', 'month',orient='h')
plt.yticks(range(0,12), [calendar.month_abbr[i] for i in range(1,13)])
g.fig.suptitle('Corridas pagas em dinheiro a cada ano.',y=1.02)
plt.legend()
for i in g.axes:
    i[0].set_yticklabels([calendar.month_abbr[i] for i in range(1,13)])
    i[0].grid()
```

## Quesito mínimo 4

Faça um gráfico de série temporal contando a quantidade de gorjetas de cada dia, nos
últimos 3 meses de 2012.

### In SQL

```python
statement = '''
SELECT DATE(pickup_datetime), COUNT('tip_amount')
FROM ride_hist
WHERE
DATE(pickup_datetime) >= "2012-10-01"
GROUP BY DATE(pickup_datetime)'''
```

```python
daily_tips_count, description = fetch_database(os.path.join('.','data','NYC_taxi.db'), statement)
print(f'SQL calculated: ')
print(daily_tips_count)
```

```python
plt.figure(figsize=(10,7))
x, y = zip(*daily_tips_count)
plt.xticks(rotation = 90)
plt.grid()
plt.plot(x,y)
plt.show()
```

## Quesito bonus 1

Qual o tempo médio das corridas nos dias de sábado e domingo;

### In SQL

```python
statement = '''
SELECT AVG(strftime('%s', dropoff_datetime) - strftime('%s', pickup_datetime))
FROM ride_hist
WHERE
strftime('%w', pickup_datetime) == '0'
OR
strftime('%w', pickup_datetime) == '6'
'''
```

```python
wkd_ride_time, description = fetch_database(os.path.join('.','data','NYC_taxi.db'), statement)
print(f'SQL calculated: ')
print(f'{wkd_ride_time[0][0]} seconds')
timedelta = pd.Timedelta(wkd_ride_time[0][0], unit = 'sec')
print(f'or {timedelta.seconds // 60:02d}:{timedelta.seconds%60}')
```

### In pandas

```python
(
    data
    .loc[
        (data['pickup_datetime'].dt.weekday == 0) | 
        (data['pickup_datetime'].dt.weekday == 6)]
    ['dropoff_datetime'] -
    data
    .loc[
        (data['pickup_datetime'].dt.weekday == 0) | 
        (data['pickup_datetime'].dt.weekday == 6)]
    ['pickup_datetime']
).mean()
```
